```python
import pandas as pd
```

```python
# Langsung baca file CSV satu per satu
df1 = pd.read_csv("../output/kemejaPria_enriched.csv")
df2 = pd.read_csv("../output/kemejaPria1_enriched.csv")

# Gabungkan kedua DataFrame (tumpuk baris)
df = pd.concat([df1, df2], ignore_index=True)

# Simpan hasil gabungan ke file baru
df.to_csv("gabungan.csv", index=False)

# Tampilkan beberapa baris pertama hasil gabungan
df.head()
```

```python
# Langkah 1: Baca file yang sudah digabung sebelumnya
df_existing = pd.read_csv("gabungan.csv")

# Langkah 2: Baca file CSV baru yang ingin digabungkan
df_new = pd.read_csv("../data/tokped_20251216-190443.csv")

# Langkah 3: Gabungkan keduanya
df_combined = pd.concat([df_existing, df_new], ignore_index=True)

# Langkah 4: Simpan hasil gabungan
df_combined.to_csv("gabungan_terbaru.csv", index=False)

# Tampilkan beberapa baris pertama hasil gabungan
df_combined.head()
```
